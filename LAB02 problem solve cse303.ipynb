```python
#Count the number of spaces in a string (use string above)
def CheckNumOfSpace(string):
    c=0
    for i in range(0,len(string)):
        if string[i]==" ":
            c=c+1
    return c
string = "Practice Problems to Drill List Comprehension in Your Head."
print("The number of spaces in a string")
print(CheckNumOfSpace(string))
```

```python
#Remove all of the vowels in a string (use string above)
def remove_vowels(string1):
    v=['a','e','i','o','u','A','E','I','O','U']
    for i in string1:
        if i not in v:
            string1=string1+i
    print(string1)

string1 = "Practice Problems to Drill List Comprehension in Your Head."
print(remove_vowels(string1))
```

```python
#5. Find all of the words in a string that are less than 5 letters (use string above)
def less_than_5words(string):
    word=string.split(" ")
    st=[]
    for i in word:
        if len(i)<5:
            st.append(i)
    return st
            
string = "Practice Problems to Drill List Comprehension in Your Head."
print(less_than_5words(string))
```

```python
#08 For all the numbers 1–1000, use a nested list/dictionary comprehension to find the highest single digit any of the numbers is divisible by
def findMax(nums):
    for num in nums:
        maximum = 0
        for i in range(1, 10):
            if (num % i) == 0:
                maximum = max(i, maximum)
        if(maximum != 0):
            print(f'index: {num}, Maximum: {maximum}')
        
nums = [i for i in range(1,1001)]
findMax(nums)
```
